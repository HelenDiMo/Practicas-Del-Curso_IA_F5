```python
#Ejemplo: En una disyunción basta que una de las proposiciones sea verdadera para que todo sea verdadero.
#Ejemplo: Voy al cine *o* me quedo en casa.
#Definicion de proposiciones.
proposicion1 = True
proposicion2 = False

#Disyuncion de las dos proposiciones
disyuncion = proposicion1 or proposicion2

print(f"Disyuncion: {disyuncion}")
```
