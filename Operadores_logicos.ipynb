OPERADORES LÓGICOS

```python
#deficion de la proposición original
calor = True

#Negación de la proposición
negacion = not calor

#imprimir los resultados
print(f"Proposición original, Hace calor: {calor}")
print(f"Negación de la proposición original, Hace calor: {negacion}")
```

```python
#Ejempplo: Hoy llueve *y* hace frío = Conjunción
#Solo será verdadera si ambas son ciertas
#Definimos las 2 proposiciones
proposicion1 = True
proposicion2 = False

#Calculamos la conjunción
conjuncion = proposicion1 and proposicion2

print(f"conjuncion: {conjuncion}")
```

```python
#Ejemplo: En una disyunción basta que una de las proposiciones sea verdadera para que todo sea verdadero.
#Ejemplo: Voy al cine *o* me quedo en casa.
#Definicion de proposiciones.
proposicion1 = True
proposicion2 = False

#Disyuncion de las dos proposiciones
disyuncion = proposicion1 or proposicion2

print(f"Disyuncion: {disyuncion}")
```
