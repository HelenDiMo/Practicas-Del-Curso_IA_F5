```python
#vector de correo electronico (vamos a filtrar correos de SPAM)
correo = ["oferta", "gratis", "dinero", "ganar", "click"]

#vector de palabras comunes en SPAM
spam_words = ["oferta", "gratis", "dinero", "ganar", "click"]

#calcular la cantidad de spam en el correo
cantidad_spam = sum([1 for word in correo if word in spam_words])

if cantidad_spam > 2 :
  print("El correo es SPAM")
else:
  print("El correo es legítimo")
```
