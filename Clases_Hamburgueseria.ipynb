```python
#este ejercicio es el que hemos hecho en clase, usando print(f) y las variables van con {}"
inventario = 100
print(f"El inventario contiene {inventario} hamburguesas")
while inventario > 0:
    if inventario <= 10:
      print("ADVENTERCIA: EL INVENTARIO ESTÁ CASI VACIO")
    num_hamburguesas = int(input("¿Cuántas hamburguesas desea comprar?"))
    if num_hamburguesas > inventario:
        print("No hay suficientes hamburguesas, solo quedan", inventario, "hamburguesas disponibles")
    else:
      inventario -= num_hamburguesas
      print(f"Has comprado", num_hamburguesas, "sólo nos quedan", inventario, "hamburguesas")

print("YA NO HAY MÁS HAMBURGUESAS")
```
