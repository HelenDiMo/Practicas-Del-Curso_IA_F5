```python
from IPython.utils.text import DollarFormatter
import pandas as pd # Importamos para manejar datos
import matplotlib.pyplot as plt # Importamos para graficar

# Cargamos la data (lo subimos en la carpetita de la izq.)
df = pd.read_csv("Unicorn_Companies.csv")

# Exploración inicial
print("----- PRIMERAS FILAS -----")
print(df.head()) # Muestra las primeras filas

print("----- INFORMACIÓN GENERAL -----")
print(df.info()) # Muestra los tipos de datos y si hay valores nulos

# Limpieza rápida - Vamos a quitar el símbolo de $ (dollar) y las comas en la columna Valuation para que sea numérica. Manejaremos los sufijos: M para Millions y B para Billions
def convert_valuation (valuation_str):
  valuation_str = valuation_str.replace('$', '').replace(',', '').strip()
  if 'B' in valuation_str:
    return float(valuation_str.replace('B', '')) * 1_000_000_000
  elif 'M' in valuation_str:
      return float(valuation_str.replace('M', '')) * 1_000_000
  else:
      return float(valuation_str)

df['Valuation'] = df['Valuation'].apply(convert_valuation)

# Estadística descriptiva
print("----- RESUMEN ESTADÍSTICO -----")
print(df.describe()) # Muestra: media, max, min...

df["Industry"].value_counts().plot(kind="bar", title="Sectores con más unicornios")

plt.xlabel("Industria")
plt.ylabel("Cantidad de Empresas")
plt.show
```
