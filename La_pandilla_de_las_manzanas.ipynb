Imagina que un grupo de amigos quiere saber más sobre su cosecha de manzanas.
Cosechan 10 manzanas y quieres analizar su peso para ver como es su producción este año.

**Queremos calcular:**

Media (promedio)

Mediana (peso central)

Moda (peso más común)

Desviación estándar (cuánto varian los pesos)

```python
import numpy as np
import statistics as stats

manzanas = [150, 180, 200, 170, 150, 190, 200, 210, 150, 180]

# Estadísticas descriptivas
media = np.mean(manzanas) # Promedio
mediana = np.median(manzanas) # Peso central
moda = stats.mode(manzanas) # Peso más común
desviacion_estandar = np.std(manzanas) # Cuánto varian los pesos

# Resultados
print(f"Análisis de la cosecha de manzanas 🍎")
print(f"Media: {media} gr")
print(f"Mediana: {mediana} gr")
print(f"Moda: {moda} gr")
print(f"Desviación estándar: {desviacion_estandar:.2f} gr") #.2f es para redondear 2 decimales.
```
