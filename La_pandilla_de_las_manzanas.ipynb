Imagina que un grupo de amigos quiere saber más sobre su cosecha de manzanas.
Cosechan 10 manzanas y quieres analizar su peso para ver como es su producción este año.

**Queremos calcular:**

Media (promedio)

Mediana (peso central)

Moda (peso más común)

Desviación estándar (cuánto varian los pesos)

```python
import numpy as np
import statistics as stats

manzanas = [150, 180, 200, 170, 150, 190, 200, 210, 150, 180]

# Estadísticas descriptivas
media = np.mean(manzanas) # Promedio
mediana = np.median(manzanas) # Peso central
moda = stats.mode(manzanas) # Peso más común
desviacion_estandar = np.std(manzanas) # Cuánto varian los pesos

# Resultados
print(f"Análisis de la cosecha de manzanas 🍎")
print(f"Media: {media} gr")
print(f"Mediana: {mediana} gr")
print(f"Moda: {moda} gr")
print(f"Desviación estándar: {desviacion_estandar:.2f} gr") #.2f es para redondear 2 decimales.
```

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

# 1. ELECCIÓN DE DATOS
print("--- 🥑 Bienvenida al Concurso de Aguacates Gigantes 🥑 ---")
bienvenida_divertida = """
      .----------.
     /    _       \\
    /   .' '.      \\
   |   /     \\      |
   |   \\     /     |   Hoy buscaremos el
    \\   '._.'     /    AGUACATE DEFINITIVO
     '-----------'
"""
print(bienvenida_divertida)
print("-" * 45)
opcion = input("¿Quieres usar datos (1) Precargados o (2) Ingresar nuevos? ")

if opcion == "1":
    aguacates = [550, 480, 520, 610, 400, 490, 505, 750, 380, 500, 510, 420]
    print("Usando datos de: la cooperativa...")
else:
    datos_user = input("Introduce los pesos separados por comas (ejemplo: 500, 600, 450): ")
    # Convertimos el texto en una lista de números
    aguacates = [float(x.strip()) for x in datos_user.split(",")]

# 2. CÁLCULOS ESTADÍSTICOS
media = np.mean(aguacates)
mediana = np.median(aguacates)
moda = stats.mode(aguacates)
desviacion = np.std(aguacates)
# Nuevo: Coeficiente de Variación
cv = (desviacion / media) * 100

# 3. IDENTIFICACIÓN
pesado = max(aguacates)
ligero = min(aguacates)

# Rango Ideal Interactivo
print("\n--- Definamos el Rango Ideal ---")
bajo = float(input("Límite inferior (ej. 400): "))
alto = float(input("Límite superior (ej. 600): "))

# Contar cuántos están en el rango
dentro_rango = [a for a in aguacates if bajo <= a <= alto]
cantidad_ideal = len(dentro_rango)

# 4. RESULTADOS POR PANTALLA
print(f"\n--- REPORTE DE COSECHA ---")
print(f"Media: {media:.2f}g | Mediana: {mediana}g | Moda: {moda}g")
print(f"Desviación Estándar: {desviacion:.2f}g")
print(f"Coeficiente de Variación: {cv:.2f}% (Menos de 15% es muy homogéneo)")
print(f"El más gordo: {pesado}g | El más flaquito: {ligero}g")
print(f"Aguacates en rango ideal ({bajo}-{alto}g): {cantidad_ideal}")

print()

# --- CÁLCULO DE ASIMETRÍA ---

from scipy.stats import skew # Importamos la herramienta para asimetría
asimetria = skew(aguacates)

print(f"Asimetría de la distribución: {asimetria:.4f}")

# 3. Interpretación didáctica
if asimetria > 0:
    print("-> La distribución es positiva: hay algunos aguacates muy pesados a la derecha.")
elif asimetria < 0:
    print("-> La distribución es negativa: hay aguacates más ligeros que 'estiran' la curva a la izquierda.")
else:
    print("-> La distribución es simétrica.")

# 5. VISUALIZACIÓN
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Histograma
ax1.hist(aguacates, bins=6, color='green', edgecolor='white')
ax1.set_title("Distribución de Pesos")
ax1.set_xlabel("Gramos")

# Boxplot (Gráfico de Caja)
ax2.boxplot(aguacates, vert=False)
ax2.set_title("Detección de Outliers (Valores Atípicos)")

plt.tight_layout()
plt.show()

print()
print()
print()

# --- GUARDAR EN ARCHIVO .TXT ---
nombre_archivo = "reporte_aguacates.txt"

# Abrimos el archivo en modo 'w' (write / escribir)
with open(nombre_archivo, "w", encoding="utf-8") as f:
    f.write("--- INFORME DEL CONCURSO DE AGUACATES ---\n")
    f.write(f"Cantidad de aguacates: {len(aguacates)}\n")
    f.write(f"Media: {media:.2f}g\n")
    f.write(f"Asimetría: {asimetria:.2f}\n")
    f.write(f"Pesos registrados: {aguacates}\n")
    f.write("----------------------------------------\n")

print(f"✅ ¡Análisis guardado con éxito en {nombre_archivo}!")
```
