Aprenderemos los conceptos básicos detras de la regresión lineal y como implementarla en Python utilizando bibliotecas basicas como: Pandas, NumPy, Matplotlib y Scikitlearn  

```python
# Importamos las bibliotecas necesarias
import pandas as pd # Libreria para manipulacion de datos
import numpy as np # Libreria para calculos matematicos
import matplotlib.pyplot as plt # Libreria para graficos
from sklearn.linear_model import LinearRegression # Modelo de regresion lineal
import warnings # Para gestionar advertencias

#ignorar todos los warnings
warnings.filterwarnings("ignore")
```

```python
# Utilizaremos un conjunto de datosque relaciona la experiencia laboral con el salario anual
# 1. Leer el archivo CSV "Salary" y cargar los datos en un DataFrame

# Pinchar en la carpetita de la izq y darle a la flechita de subir para cargar el archivo - se tiene que subir todo el rato, porque se va a borrar una vez salgamos de aqui.

df = pd.read_csv("Salary.csv")

# Mostrar las primeras 6 columnas que tiene el DataFrame para revisar los datos.
df.head(6)
```

Graficamos los datos para observar cuál es la correlación entre las variables.

```python
# Creamos un gráfico de dispersión
plt.scatter(df["YearsExperience"], df["Salary"])

# Vamos a agregar etiquetas al eje X y al eje Y
plt.xlabel("Años de Experiencia")
plt.ylabel("Salario Anual")

# Mostramos el gráfico
plt.show()
```

```python
# Verificamos que la correlacion es positiva y lineal entre las variables, usamos un modelo de regresion lineal.

# Creamos un objeto de regresion linal
modelo = LinearRegression()

# Ajustamos el modelo utilizando los datos de Years Experience como características y Salary como variable.

modelo.fit(df[["YearsExperience"]], df["Salary"]) # Years Experience lleva doble [[]] porque es una característica

# Vamos a realizar una predicción anual para 15 años de experiencia
prediccion = modelo.predict([[15]])

print(f"Predicción salarial anual para 15 años de experiencia: {prediccion}")
```

```python
# Creamos el gráfico de dispersión de los ejes
plt.scatter(df["YearsExperience"], df["Salary"])

# Superponer la línea de regresion del modelo en el gráfico
plt.plot(df["YearsExperience"], modelo.predict(df[["YearsExperience"]]), color="red")

# Agregamos etiquetas
plt.xlabel("Años de Experiencia")
plt.ylabel("Salario Anual")

plt.show()
```

Vamos a usar una metrica de evaluación para ver si el modelo funciona o no, como el coeficiente de determinación (R^2)

```python
# Imprimimos el coeficiente de determinación que indica cuánta varianza existe en la variable dependiente Salario
print(f"Coeficiente de determinación: {modelo.score(df[["YearsExperience"]], df["Salary"])}")
print(f"R^2: {modelo.score(df[["YearsExperience"]], df["Salary"])}")
```
